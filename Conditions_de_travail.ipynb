L’objet du projet est d’analyser, à partir de l’enquête Conditions de Travail 2013, les déterminants pour qu’une
femme soit le principal apporteur de ressources actuel du ménage. 

```python
import pandas as pd
```

```python
df= pd.read_sas("/Users/paulaneira/Desktop/Project 3 statistics/individus_ct2013 (1).sas7bdat", format="sas7bdat", index=None, encoding="latin-1")
df
```

```python
expected_number_of_columns = len(df.columns)
expected_number_of_columns
```

```python
chunk_size = 1000 

problematic_rows = []

for chunk in pd.read_sas("/Users/paulaneira/Desktop/Project 3 statistics/individus_ct2013 (1).sas7bdat", format="sas7bdat", chunksize=chunk_size):
    for index, row in chunk.iterrows():
        if len(row) != expected_number_of_columns:
            problematic_rows.append((index, row))

if problematic_rows:
    print(f"Found {len(problematic_rows)} problematic rows with column count mismatch.")
    for row in problematic_rows[:10]:  # Print first 10 problematic rows
        print(f"Row {row[0]}: {len(row[1])} columns")
else:
    print("No column count mismatches found.")
```
