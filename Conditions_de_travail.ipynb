```python

# Function to convert values
def check_input(input_value):
    if isinstance(input_value, (int, float)):
        return input_value
    elif isinstance(input_value, str):
        try:
            num_value = float(input_value)
            if num_value != num_value:  # Check if it's NaN (since NaN != NaN)
                return 0
            return num_value
        except ValueError:
            return 0
    elif isinstance(input_value, bool):
        return 0
    else:
        return 0


# Apply the function to the column
#df_filtered['DEPNAIS'] = df_filtered['DEPNAIS'].fillna(0)
#df_filtered['DEPNAIS'] = df_filtered['DEPNAIS'].apply(check_input)
# Fill NaN values with 0 (if needed)
df_filtered['DEPNAIS'] = df_filtered['DEPNAIS'].fillna(0)

# Apply the function to the column
df_filtered['DEPNAIS'] = df_filtered['DEPNAIS'].apply(check_input)
```

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

# Supposons que nous souhaitons prédire 'emploi_avec'
y = df_filtered['variable_binaire_dependant']
X = df_filtered.drop(columns=['variable_binaire_dependant',"SEXE","PRACT"])

# Ajouter une constante pour le modèle
X = sm.add_constant(X)
```

```python
model = sm.Logit(y, X)
result = model.fit()
print(result.summary())
```

```python
# Prédire les probabilités
df['predicted_prob'] = result.predict(X)

# Tracer les probabilités prédites
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='predicted_prob', data=df, hue='sexe_M', style='emploi_sans')
plt.xlabel('Âge')
plt.ylabel('Probabilité prédite d\'emploi_avec')
plt.title('Probabilité prédite d\'emploi_avec en fonction de l\'âge')
plt.show()









```
